```python
#Load the Data
import pandas as pd

# Load the CSV file
file_path = 'C:/repos/group_project_1/Ruby_Project/Electric_Vehicle_Population_Size_History_By_County.csv'
ev_data = pd.read_csv(file_path)

# Convert the 'Date' column to datetime format
ev_data['Date'] = pd.to_datetime(ev_data['Date'])

# Display the first few rows
ev_data.head()

#Aggregate Data by Date
# Aggregate data by date
agg_data = ev_data.groupby('Date').agg({
    'Battery Electric Vehicles (BEVs)': 'sum',
    'Plug-In Hybrid Electric Vehicles (PHEVs)': 'sum',
    'Electric Vehicle (EV) Total': 'sum',
    'Non-Electric Vehicle Total': 'sum',
    'Total Vehicles': 'sum'
}).reset_index()
#Visualize the Trends
import matplotlib.pyplot as plt
import seaborn as sns

# Set the style for the plots
sns.set(style="whitegrid")

# Plot the trends of BEVs, PHEVs, and Non-EVs over time
plt.figure(figsize=(8, 4))
sns.lineplot(data=agg_data, x='Date', y='Battery Electric Vehicles (BEVs)', label='BEVs')
sns.lineplot(data=agg_data, x='Date', y='Plug-In Hybrid Electric Vehicles (PHEVs)', label='PHEVs')
sns.lineplot(data=agg_data, x='Date', y='Non-Electric Vehicle Total', label='Non-EVs')
plt.title('Trends of Electric and Non-Electric Vehicles Over Time')
plt.xlabel('Date')
plt.ylabel('Number of Vehicles')
plt.savefig('Graph_1')
plt.legend()
plt.show()

#Calculate and Compare Adoption Rates
# Calculate the percentage of electric vehicles
agg_data['Percent Electric Vehicles'] = (agg_data['Electric Vehicle (EV) Total'] / agg_data['Total Vehicles']) * 100

# Plot the percentage of electric vehicles over time
plt.figure(figsize=(8, 4))
sns.lineplot(data=agg_data, x='Date', y='Percent Electric Vehicles')
plt.title('Percentage of Electric Vehicles Over Time')
plt.xlabel('Date')
plt.ylabel('Percentage of Electric Vehicles')
plt.savefig('Graph_2')
plt.show()

#Geographic Analysis
# Aggregate data by county and state
county_data = ev_data.groupby(['County', 'State']).agg({
    'Battery Electric Vehicles (BEVs)': 'sum',
    'Plug-In Hybrid Electric Vehicles (PHEVs)': 'sum',
    'Electric Vehicle (EV) Total': 'sum',
    'Non-Electric Vehicle Total': 'sum',
    'Total Vehicles': 'sum'
}).reset_index()

# Calculate the percentage of electric vehicles for each county
county_data['Percent Electric Vehicles'] = (county_data['Electric Vehicle (EV) Total'] / county_data['Total Vehicles']) * 100

# Display the top 10 counties with the highest percentage of electric vehicles
top_counties = county_data.sort_values(by='Percent Electric Vehicles', ascending=False).head(10)
top_counties
```

```python
#Electric cars and non electric cars Data for Washington State
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the CSV file
file_path = 'C:/repos/group_project_1/Ruby_Project/Electric_Vehicle_Population_Size_History_By_County.csv'
ev_data = pd.read_csv(file_path)

# Convert the 'Date' column to datetime format
ev_data['Date'] = pd.to_datetime(ev_data['Date'])

# Filter data for Washington state
wa_data = ev_data[ev_data['State'] == 'WA']

# Aggregate data by date for Washington state
wa_agg_data = wa_data.groupby('Date').agg({
    'Battery Electric Vehicles (BEVs)': 'sum',
    'Plug-In Hybrid Electric Vehicles (PHEVs)': 'sum',
    'Electric Vehicle (EV) Total': 'sum',
    'Non-Electric Vehicle Total': 'sum',
    'Total Vehicles': 'sum'
}).reset_index()

# Set the style for the plots
sns.set(style="whitegrid")

# Plot the trends of BEVs, PHEVs, and Non-EVs over time in Washington state
plt.figure(figsize=(8, 4))
sns.lineplot(data=wa_agg_data, x='Date', y='Battery Electric Vehicles (BEVs)', label='BEVs')
sns.lineplot(data=wa_agg_data, x='Date', y='Plug-In Hybrid Electric Vehicles (PHEVs)', label='PHEVs')
sns.lineplot(data=wa_agg_data, x='Date', y='Non-Electric Vehicle Total', label='Non-EVs')
plt.title('Trends of Electric and Non-Electric Vehicles Over Time in Washington State')
plt.xlabel('Date')
plt.ylabel('Number of Vehicles')
plt.legend()
plt.show()

# Calculate the percentage of electric vehicles in Washington state
wa_agg_data['Percent Electric Vehicles'] = (wa_agg_data['Electric Vehicle (EV) Total'] / wa_agg_data['Total Vehicles']) * 100

# Plot the percentage of electric vehicles over time in Washington state
plt.figure(figsize=(8, 4))
sns.lineplot(data=wa_agg_data, x='Date', y='Percent Electric Vehicles')
plt.title('Percentage of Electric Vehicles Over Time in Washington State')
plt.xlabel('Date')
plt.ylabel('Percentage of Electric Vehicles')
plt.savefig('Graph_3')
plt.show()
```

```python
# Growth of non-electric vehicles in Washington State
# Filter the data for Washington and extract the relevant columns

import pandas as pd

# Load the CSV file
file_path = 'C:/repos/group_project_1/Ruby_Project/Electric_Vehicle_Population_Size_History_By_County.csv'
data = pd.read_csv(file_path)

# Filter the dataset for Washington State
wa_data = data[data['State'] == 'WA']

# Select relevant columns for analysis
wa_data = wa_data[['Date', 'County', 'Non-Electric Vehicle Total']]

# Convert 'Date' column to datetime format for time series analysis
wa_data['Date'] = pd.to_datetime(wa_data['Date'])

 # Aggregate data by date to get the total number of non-electric vehicles in Washington State over time
wa_aggregated_data = wa_data.groupby('Date')['Non-Electric Vehicle Total'].sum().reset_index()

# Display the first few rows of the aggregated dataset
wa_aggregated_data.head()

import matplotlib.pyplot as plt

# Plot the growth of non-electric vehicles in Washington State over time
plt.figure(figsize=(8, 4))
plt.plot(wa_aggregated_data['Date'], wa_aggregated_data['Non-Electric Vehicle Total'], marker='o', linestyle='-')
plt.xlabel('Date')
plt.ylabel('Total Number of Non-Electric Vehicles')
plt.savefig('Graph_4')
plt.title('Growth of Non-Electric Vehicles in Washington State Over Time')
plt.grid(True)
plt.show()
```

```python
#Compare the growth of electric vehicles (EVs) and non-electric vehicles in Washington State

import pandas as pd
# Load the CSV file
file_path = 'C:/repos/group_project_1/Ruby_Project/Electric_Vehicle_Population_Size_History_By_County.csv'
data = pd.read_csv(file_path)

# Filter the dataset for Washington State and create a copy to avoid the warning
wa_data = data[data['State'] == 'WA'].copy()

# Convert 'Date' column to datetime format
wa_data['Date'] = pd.to_datetime(wa_data['Date'])

# Aggregate data by date to get the total number of electric vehicles
wa_ev_data = wa_data.groupby('Date')['Electric Vehicle (EV) Total'].sum().reset_index()

# Aggregate data by date to get the total number of non-electric vehicles
wa_non_ev_data = wa_data.groupby('Date')['Non-Electric Vehicle Total'].sum().reset_index()

import matplotlib.pyplot as plt

# Plot the growth of electric and non-electric vehicles in Washington State over time
plt.figure(figsize=(8, 4))
plt.plot(wa_ev_data['Date'], wa_ev_data['Electric Vehicle (EV) Total'], marker='o', linestyle='-', label='Electric Vehicles')
plt.plot(wa_non_ev_data['Date'], wa_non_ev_data['Non-Electric Vehicle Total'], marker='o', linestyle='-', label='Non-Electric Vehicles')
plt.xlabel('Date')
plt.ylabel('Total Number of Vehicles')
plt.title('Growth of Electric and Non-Electric Vehicles in Washington State Over Time')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Groth of electric cars in the washington state
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file
file_path = 'C:/repos/group_project_1/Ruby_Project/Electric_Vehicle_Population_Size_History_By_County.csv'
data = pd.read_csv(file_path)

# Filter the dataset for Washington State and create a copy to avoid the warning
wa_data = data[data['State'] == 'WA'].copy()

# Convert 'Date' column to datetime format
wa_data['Date'] = pd.to_datetime(wa_data['Date'])

# Filter the data for the last five years
last_five_years = wa_data[wa_data['Date'] >= (pd.to_datetime('today') - pd.DateOffset(years=5))]

# Aggregate data by date to get the total number of electric vehicles
wa_ev_data = last_five_years.groupby('Date')['Electric Vehicle (EV) Total'].sum().reset_index()

# Create the bar graph
plt.figure(figsize=(8, 4))
plt.bar(wa_ev_data['Date'], wa_ev_data['Electric Vehicle (EV) Total'], color='skyblue')
plt.xlabel('Date')
plt.ylabel('Total Number of Electric Vehicles')
plt.title('Growth of Electric Vehicles in Washington State Over the Last Five Years')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()
```

```python
import pandas as pd

# Load the data
file_path =  'C:/repos/group_project_1/Ruby_Project/Electric_Vehicle_Population_Size_History_By_County.csv'
df = pd.read_csv(file_path)

# Filter the data for Washington state only
wa_data = df[df['State'] == 'WA']

# Convert the 'Date' column to datetime format
wa_data['Date'] = pd.to_datetime(wa_data['Date'])

# Extract the year from the 'Date' column
wa_data['Year'] = wa_data['Date'].dt.year

# Group the data by year and sum the relevant columns
annual_data = wa_data.groupby('Year').sum().reset_index()

# Plot the data
import matplotlib.pyplot as plt


# Plotting the trends of electric and non-electric vehicles over the past five years
plt.figure(figsize=(8, 4))
plt.bar(annual_data['Year'] - 0.2, annual_data['Electric Vehicle (EV) Total'], width=0.4, label='Electric Vehicles', color='blue')
plt.bar(annual_data['Year'] + 0.2, annual_data['Non-Electric Vehicle Total'], width=0.4, label='Non-Electric Vehicles', color='orange')
plt.xlabel('Year')
plt.ylabel('Number of Vehicles')
plt.title('Trends of Electric and Non-Electric Vehicles in Washington State (Past Five Years)')
plt.legend()
plt.show()
```
